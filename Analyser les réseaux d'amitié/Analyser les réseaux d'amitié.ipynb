Les « résultats scolaires » sont le niveau d'acquisition par les élèves de connaissances, de compétences et d'aptitudes par rapport aux objectifs spécifiés dans le programme éducatif, de l'école primaire à l'université.

La recherche des facteurs qui influencent le rendement scolaire est essentielle pour comprendre ce qui contribue à la réussite ou à l’échec des étudiants. De nombreuses études ont été menées pour identifier les facteurs qui influent sur les résultats d'apprentissage des étudiants, notamment :

La recherche de Kim (2015) sur les effets des activités de discussion, de l'interaction avec les conférenciers et de l'apprentissage autodirigé sur les résultats d'apprentissage des étudiants. L'objectif de cette étude était de confirmer la relation structurelle entre les activités de discussion, l'interaction avec les professeurs, l'apprentissage autodirigé et les résultats scolaires des étudiants de l'enseignement supérieur. Les données ont été recueillies dans le cadre de l'évaluation nationale de l'engagement des étudiants dans l'apprentissage (National Assessment of Student Engagement in Learning, NASEL). NASEL a été développé dans le cadre de la recherche stratégique sur l'amélioration de la qualité de l'enseignement et de l'apprentissage universitaires dans le cadre du plan quinquennal (2013-2017) de l'Institut coréen pour le développement de l'éducation Quoc. Cette étude utilise les données de l'institution de l'auteur provenant d'une enquête nationale de 2013 sur les processus d'enseignement et d'apprentissage des étudiants coréens, à laquelle ont participé 28 095 étudiants de 47 établissements d'enseignement supérieur de quatre ans. L'échantillon comprenait 336 répondants à l'enquête. L'analyse statistique a démontré que les interactions avec le corps enseignant et l'apprentissage autonome influencent directement les résultats de l'apprentissage. Toutefois, les activités de discussion n'ont pas eu d'impact significatif sur les résultats d'apprentissage, sauf si la capacité d'apprentissage autonome a été utilisée comme paramètre. De même, les interactions avec les instructeurs ont un impact indirect sur les résultats d'apprentissage en utilisant la capacité d'apprentissage autonome comme paramètre. Sur la base de ces résultats, l'auteur a également proposé un certain nombre d'outils fonctionnels pour améliorer les résultats d'apprentissage.

L'étude intitulée "Origine migratoire et inégalités scolaires : une analyse longitudinale des performances académiques des descendants d'immigrés en France et en Angleterre" examine comment l'origine migratoire façonne la stratification éducative parmi les groupes d'immigrés dans ces deux pays. Elle met en évidence une diversité significative dans les résultats éducatifs entre différents groupes migrants, notant en particulier la supériorité académique constante des descendants d'immigrés asiatiques, en particulier ceux d'origine chinoise en Angleterre. En outre, le texte souligne que les conditions éducatives en Angleterre sont généralement plus favorables pour les descendants d'immigrés que celles en France, où l'écart de performance entre les autochtones et les descendants d'immigrés s'élargit, souvent au détriment de ces derniers. Le document explore la possibilité que les progrès des descendants d'immigrés dans ces pays soient liés à leur développement linguistique précoce, bien que cela ne puisse être affirmé avec certitude faute de données suffisantes. Il cite également des études menées en France, en Grande-Bretagne et aux États-Unis qui soutiennent cette hypothèse. En France, la ségrégation sociale et ethnique dans les écoles secondaires semble nuire à la progression académique des descendants d'immigrés, tandis qu'en Angleterre, la concentration ethnique dans les écoles ne semble pas impacter négativement leurs résultats scolaires. Enfin, l'étude analyse les stratégies sociologiques adoptées par les immigrants pour améliorer les performances scolaires de leurs enfants, indépendamment de leur propre statut social et éducatif. Elle observe que les immigrants ayant un niveau d'éducation et un statut socio-économique supérieurs (« sélectionnés positivement ») dans leur pays d'origine ont généralement des attentes plus élevées pour la réussite scolaire de leurs enfants. Cette recherche fournit une perspective longitudinale sur la stratification des performances académiques des descendants d'immigrés en France et en Angleterre, de l'école primaire au lycée. L'étude se concentre sur l'impact de l'environnement collégial et de l'engagement en classe sur les résultats scolaires, ainsi que sur les effets médiateurs de l'intégration des expériences collégiales des étudiants. Utilisant les données de 18 551 étudiants de 32 universités coréennes, cette étude a utilisé la modélisation d’équations structurelles pour étudier les relations structurelles entre les variables. Les auteurs ont constaté que l’environnement universitaire, l’interaction étudiants-professeurs et la participation en classe ont des impacts significatifs sur l’amélioration des performances académiques des étudiants et sont indirectement liés aux résultats d’apprentissage grâce à l’intégration expérientielle. D'après la présentation du codebooks, la variable « grade » dans les données PACHA représente une note standardisée de la note moyenne obtenue par un étudiant au cours du trimestre en cours. La note moyenne (moyenne) de tous les élèves de l'école est soustraite de la note moyenne de chaque élève. Cette étape décale la répartition des notes pour que le centre soit à zéro. Après avoir soustrait la moyenne, le résultat est divisé par l'écart type des notes de l'école. L'écart type est une mesure de la répartition des notes. Cette étape met à l'échelle la distribution afin qu'elle représente le nombre d'écarts types par rapport à la moyenne de la note de chaque élève.


En conséquence, une valeur « grade » :

•	Autour de 0 indique une note moyenne proche de la moyenne scolaire.

•	Un résultat positif indique les élèves ayant une note standardisée positive obtiennent des résultats supérieurs à la moyenne de l'école.

•	Un résultat négatif indique les élèves ayant une note standardisée négative ont des résultats inférieurs à la moyenne de l'école.

Cette standardisation permet de comparer les notes de différentes classes ou matières avec des échelles de notation potentiellement différentes, en se concentrant sur les performances relatives des élèves au sein de la population scolaire.

```python
file_path_network1 = '/content/drive/MyDrive/Stats/allf_school1_w6.csv'
network_data = pd.read_csv(file_path_network1)

network_data.set_index('Unnamed: 0', inplace=True)
G = nx.from_pandas_adjacency(network_data)

# Drawing the network
plt.figure(figsize=(12, 12))
nx.draw_networkx(G, node_size=50, with_labels=False)
plt.title("Réseaux d'ami.es et de très bon.nes ami.es")
plt.show()
```

```python
# Calculating network density
density = nx.density(G)

# Identifying communities using the Clauset-Newman-Moore community detection algorithm
from networkx.algorithms import community
communities = community.greedy_modularity_communities(G)

# Calculating degree centrality for central vs peripheral students
degree_centrality = nx.degree_centrality(G)


density, len(communities), degree_centrality
```

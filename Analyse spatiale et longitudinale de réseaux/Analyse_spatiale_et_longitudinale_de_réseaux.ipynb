1. **Introduction**

L'analyse des réseaux personnels est devenue un domaine d'étude important pour comprendre non seulement comment les individus se connectent et interagissent mais aussi comment ces interactions façonnent les communautés, les organisations et même les associations entières en général. Le Panel de Caen offre une opportunité unique d’explorer ces dynamiques. À l’aide d’un ensemble de données détaillées collectées sur une longue période de temps, cette étude vise à décrypter les modèles complexes et les tendances de développement des réseaux personnels, en mettant en évidence la nature dynamique et en constante évolution des relations entre les personnes. À travers cette analyse, nous cherchons à mieux comprendre comment les relations sociales se forment, se développent et influencent la cohésion et la diversité des réseaux personnels.

La recherche et l'analyse des réseaux personnels ont prospéré dans l'anthropologie et la sociologie occidentales dans les années 1980, avec les auteurs éminents Bourdieu et Coleman. Bourdieu a souligné que le réseau personnel d'un individu dépend de la nature et de l'étendue de ses relations et de la quantité de différents types de capital (économique, culturel, symbolique, etc.) que chaque personne dans son réseau possède. Selon Bourdieu, ces ressources sociales et symboliques sont essentielles pour mobiliser les contacts nécessaires afin d'agir efficacement dans la société. En d'autres termes, l'accès à un réseau personnel étendu et diversifié, composé de personnes dotées de diverses ressources, est considéré comme un avantage pour un acteur particulier (Bourdieu, 1986). Pendant ce temps, Coleman soutient que les réseaux personnels, les normes et la confiance dans la société sont ce qui permet aux membres d'agir ensemble efficacement pour atteindre leurs objectifs communs (Coleman, 1988).

Dans le même temps, le point de vue de deux chercheurs Wellman et Berkowtz a également déclaré que le réseau personnel est le concept de nœuds et de connexions entre ces nœuds. Les nœuds ici sont les individus qui agissent et les connexions sont les relations entre ces individus. En d'autres termes, un réseau personnel se compose de membres individuels et de connexions entre les membres individuels (Wellman et Berkowtz, 1988).

À la fin des années 1990, Putnam - un politologue américain a continué à prêter attention et à développer ses opinions sur les réseaux personnels. Selon lui : « Les caractéristiques de la vie sociale - les réseaux, les normes et les croyances sont celles qui permettent aux gens d'agir ensemble plus efficacement dans la poursuite d'objectifs communs. » (Putnam, 1995).

On peut dire que les chercheurs occidentaux ne sont pas seulement crédités d'avoir initié la tendance de la recherche, en appliquant la perspective des réseaux personnels pour analyser les relations d'échange et les réseaux personnels, mais aussi en soulignant les effets positifs des réseaux personnels et les effets négatifs des interactions dans les réseaux personnels sur le développement personnel et social.

Dans la cadre de la recherche sur les réseaux personnels, Michel Grossetti a mené une étude comparative entre Toulouse et celle réalisée par Claude S. Fischer dans le nord de la Californie. Fischer avait mené une enquête à grande échelle sur les réseaux personnels dans le but de tester la théorie selon laquelle l'urbanisation massive avait entraîné un déclin des liens communautaires. Cette étude avait démontré que les opportunités de soutien social n'avaient pas diminué dans les contextes urbains ou ruraux. Les études menées par Fischer et Wellman sont devenues des références classiques pour la recherche sur les réseaux personnels. De son côté, l'étude de Michel Grossetti à Toulouse visait à comparer les résultats obtenus avec ceux de Fischer et à déterminer si les réseaux français diffèrent de ceux des autres régions. Les résultats ont montré que les réseaux urbains adoptent une structure légèrement différente, en particulier avec une densité plus faible. Bien que les réseaux français soient légèrement plus grands, ils sont similaires en termes d'hétérogénéité. Dans l'ensemble, ces études ont contribué à notre compréhension du fonctionnement des réseaux personnels et de la manière dont ils sont influencés par différents contextes. (Grossetti, 2007)

Un aspect tout aussi important mais moins étudié du domaine des réseaux est le rôle et l'effet de l'absence totale de réseau. En demandant aux répondants de citer les noms des personnes avec lesquelles ils entretiennent certaines relations ou activités, l'enquêteur peut déterminer si une personne a ou non des relations environnantes. L'augmentation de l'espérance de vie pose des questions sur la qualité de vie des personnes âgées. En effet, nombre d'événements peuvent bouleverser la vie d'une personne âgée et fragiliser ses habitudes de vie, ses relations (Campéon, 2016). Il existe 8 facteurs importants étroitement liés à la solitude comme : la retraite, le changement d'état civil vers le veuvage ou le décès d'un proche, le sentiment de solitude, la fréquence des contacts avec les enfants, des visites, la prise de médicaments de soutien, la situation financière, la survenue d'états dépressif au cours des quatre dernières semaines. (Croutte, 2004). De plus, l'affaiblissement de la relation conduit à l'isolement, des ressources psychiques, la perte, la faiblesse des réseaux personnels contribuent au deuil compliqué ou un deuil pathologique des personnes âgées (Josse, 2020).

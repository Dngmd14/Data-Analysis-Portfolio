In today's digital age, social networks have become an indispensable part of our lives. Social media influencers (SMI) are increasingly influencing users' awareness, behavior and lifestyle. One of the prominent trends is the use of communication strategies and techniques by SMIs to convey healing content, to help viewers cope with psychological problems and improve mental health.

The study was to determine the role and effectiveness of communication strategies and techniques used by SMIs in delivering healing content, and how demographic and sociodemographic factors influence their reception. Viewer capture and interaction with this content.

**Research Objectives**


*   Analyze the communication strategies and techniques that SMIs use to deliver healing content.
*   Assess the impact of demographic factors (age, gender, marital status, education level, mental health) on uptake and engagement with healing content.
*   Identify the key factors in building trust and positive interactions between SMI and audiences.

This research is based in Vietnam, part of the Professional Practice project. The research was conducted in collaboration with VU Quang Huy - Student of University of Technology Sydney.

**DEFINITIONS**

- Social media influencers are social network users who have great influence on one or more professional and/or life aspects. Their content and opinions have great influence, can reach many other social network users and receive their exchange, discussion or support. The most popular examples of these people are: YouTubers, TikTokers, etc. Some popular social media platforms on which they are active: Facebook, Instagram, YouTube, TikTok, blogging platforms, etc.

- Healing content or self-healing content (healing/self-healing content) is content shared on social networks (can be in the form of articles, photos, videos, podcasts, etc.) that can The ability to help viewers relieve stress and psychological pressure. This content may include (but is not necessarily limited to) content that shows sharing, sympathy, encouragement, and encouragement for viewers in many aspects, helping viewers strengthen their mental stability. spirit, feel more secure, comforted and understood.

- ASMR content (Autonomous Sensory Meridian Response, roughly translated as "Independent Orgasmic Response"): content that causes a shivering response in the head or neck after receiving certain stimuli such as soft sounds A pattern or sound that repeats periodically.

The agents that create the ASMR effect are very diverse and are quite familiar in life, from tangible to invisible objects. ASMR can come from sounds, images, scents, textures or touches, etc.

The most common ASMR effects usually come from sound. But this sound must have a special impact on the human touch, creating a feeling of flying and comfort, which can be:

+ Someone's whisper
+ Light tapping or scratching sounds on surfaces: Video makers often lightly tap on tables, books and glasses of water or lightly scratch on the mic
+ Book flipping sound:
+ The sound of rain, the sound of pouring water, the sound of chewing food, the sound of dry leaves rustling,...

```python
import statsmodels.api as sm

# Filter the dataset for records where Q1.1 is 2 (those who have watched self-healing content)
filtered_df = df[df['Q1.1'] == 2]

# Select relevant columns: Q1.3a, Q2.1a, Q2.1b, Q2.1c, Q2.1d
regression_df = filtered_df[['Q1.3a', 'Q2.1a', 'Q2.1b', 'Q2.1c', 'Q2.1d']]

# Drop rows with any missing values
regression_df = regression_df.dropna()

# Define the dependent variable (Q1.3a) and independent variables (Q2.1a, Q2.1b, Q2.1c, Q2.1d)
Y = regression_df['Q1.3a']
X = regression_df[['Q2.1a', 'Q2.1b', 'Q2.1c', 'Q2.1d']]

# Add a constant to the independent variables
X = sm.add_constant(X)

# Perform the linear regression
model = sm.OLS(Y, X).fit()

# Get the summary of the regression model
regression_summary = model.summary()

regression_summary
```

```python
# Extracting coefficients, standard errors, t-values, and p-values into a dataframe
results_df = pd.DataFrame({
    'Coefficient': model.params,
    'Std. Error': model.bse,
    't-value': model.tvalues,
    'P>|t|': model.pvalues,
    '[0.025': model.conf_int()[0],
    '0.975]': model.conf_int()[1]
})

print(results_df)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load the dataset
file_path = 'drive/MyDrive/3b.-Data-Encoding-finalised.xlsx'
data_encoding_df = pd.read_excel(file_path)

# Filter the dataset for records where Q1.1 is 2 (those who have watched self-healing content)
filtered_df = data_encoding_df[data_encoding_df['Q1.1'] == 2]

# Rename columns to replace periods with underscores
filtered_df.columns = filtered_df.columns.str.replace('.', '_')

# Define the demographic variables
demographic_vars = ['Q3_1', 'Q3_2', 'Q3_3', 'Q3_4', 'Q3_7a', 'Q3_7b', 'Q3_7c']

# Define the dependent variables related to content creator opinions
dependent_vars = ['Q2_4a', 'Q2_4b', 'Q2_4c', 'Q2_4d', 'Q2_4e', 'Q2_4f', 'Q2_4g', 'Q2_4h']

# Create an empty DataFrame to store ANOVA results
anova_results = pd.DataFrame(columns=['Dependent Variable', 'Demographic Variable', 'F-Statistic', 'p-Value'])

anova_results_list = []

# Perform ANOVA for each combination of demographic and dependent variable
for dep_var in dependent_vars:
    for demo_var in demographic_vars:
        # Ensure the demographic variable is categorical
        filtered_df[demo_var] = filtered_df[demo_var].astype('category')
        formula = f'{dep_var} ~ C({demo_var})'
        model = ols(formula, data=filtered_df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        anova_results_list.append({
            'Dependent Variable': dep_var,
            'Demographic Variable': demo_var,
            'F-Statistic': anova_table['F'][0],
            'p-Value': anova_table['PR(>F)'][0]
        })

# Convert the list of results to a DataFrame
anova_results = pd.DataFrame(anova_results_list)

# Filter results for significant p-values
significant_results = anova_results[anova_results['p-Value'] < 0.05]

significant_results
```

Dependent Variable: These are the variables where we are interested in the influence of demographic variables. For example, focus on arguments and evidence, compare with personal opinions, seek additional information to ensure accuracy, etc.

Demographic Variable: These are demographic factors such as education level, gender, and mental health status.
